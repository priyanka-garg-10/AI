While loop

```python
# Loop Control statment 
count=1
while(count>0):
    if count==3:
        break
    print(count)
    count+=1
```

```python
# Loop Control statment 
count=1
while(count>0 and count<6):
    if count==3:
        count+=1
        continue
    print(count)
    count+=1
```
