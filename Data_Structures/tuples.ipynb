Tuples are orderder collection of elements that is immutable

```python
t=()
print(t)
print(type(t))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6])
numbers
```

```python
## Accessing Tuple Elements

print(numbers[2])
print(numbers[-1])

#slicing method same as list 
```

```python
## Tuple Operations

numbers =(1,2,3,4,5)
mixed=("Priyanka",7,True)
concatenation_tuple=numbers + mixed
print(concatenation_tuple)

mixed*3
```

Immutable Nature Of Tuples

```python
## Tuples are immutable, meaning their elements cannot be changed once assigned.

lst=[1,2,3,4,5]
print(lst)

lst[1]="Priyanka"
print(lst)
```

```python
numbers[1]="Priyanka"
```

Tuple Methods


```python
print(numbers.count(1))
print(numbers.index(3))
```

Packing and Unpacking tuple

```python
packed_tuple=1,"Hello",3.14
print(packed_tuple)

##unpacking a tuple
a,b,c=packed_tuple

print(a)
print(b)
print(c)

## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```
