Lamba functions are small anonymous funtions defined using the lambda keyword.They can have any number of argument but only one expression.They are commonly used for short operation or as a agrument to pass in highed order funtion

```python
lambda arguments: expression
```

```python
addition=lambda a,b:a+b
type(addition)
print(addition(5,6))
```

```python
def even(num):
    if num%2==0:
        return True
    
even(24)
```
